```python
lon = matrix['LON']
lat = matrix['LAT']
ref = matrix['REFLECT']
```

```python
import numpy as np
import matplotlib.pyplot as plt

plt.scatter(lon, lat, c=ref, s=350, alpha=0.5, cmap='viridis', edgecolors='b', label='Reflect')
plt.title('Cloud')
```
